```python
import os
import cv2

DATASET_PATH = "data/raw"
classes = os.listdir(DATASET_PATH)
corrupted = []

for cls in classes:
    class_dir = os.path.join(DATASET_PATH, cls)
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        img = cv2.imread(img_path)
        if img is None:
            corrupted.append(img_path)
            os.remove(img_path)  # Delete corrupted file

print(f"Removed {len(corrupted)} corrupted images.")
```

```python
DATASET_PATH = "../data/raw"   # Folder with subfolders for each class
IMG_SIZE = (224, 224)       # Chosen resolution
USE_GRAYSCALE = False        # True for grayscale, False for RGB
```

```python
# --------------------------
# Load and Preprocess Function
# --------------------------
def preprocess_image(image_path, img_size=IMG_SIZE, grayscale=USE_GRAYSCALE):
    img = cv2.imread(image_path) # computer vision
    if img is None:
        return None
    
    
    img = cv2.resize(img, img_size) # 224 * 224
    
    # Convert to grayscale if needed
    if grayscale:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Normalize pixels to 0-1
    img = img / 255.0
    
    return img
```
